```python
# query gnd resources
sparql = SPARQLWrapper("http://data.judaicalink.org/sparql/query")
sparql.setQuery("""
    PREFIX owl: <http://www.w3.org/2002/07/owl#>
    PREFIX gnd: <http://d-nb.info/gnd/>
    PREFIX gndo: <http://d-nb.info/standards/elementset/gnd#>
    PREFIX dnbt: <http://d-nb.info/standards/elementset/dnb#>
    SELECT ?gnd_res ?gnd_id
    WHERE {
        GRAPH <http://data.judaicalink.org/data/ep> {
            ?ep owl:sameAs ?gnd_res 
        }
        ?gnd_res gndo:gndIdentifier ?gndid 
        }
""")
sparql.setReturnFormat(JSON)
results = sparql.query().convert()
```

```python
gnd_resources = set([])
for res in results['results']['bindings']:
    gnd_res = res['gnd_res']['value']
    gnd_resources.add(gnd_res)
```
