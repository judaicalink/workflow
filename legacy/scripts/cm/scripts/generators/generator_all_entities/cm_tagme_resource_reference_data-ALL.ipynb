This script generates:  
1) URIs for resources, i.e. entities found in Compact Memory by Tagme _independently from their availability in Judaicalink_.

```python
import os, json, pickle
import urllib.parse
```

```python
def generate_res_ref_data(cm_tagme_mentions, link_prob=0.5, rho=0.3, save=False, out_path=""):
    
    print("Generating resource and reference data with link probability={} and rho={}...".format(link_prob, rho))
    
    base_dbp = 'http://de.dbpedia.org/resource/'
    base_ref = 'http://data.judaicalink.org/data/cm-tagme/'

    cm_mentions = []
    i = 0
    ref_index = 10000000
    for mention in cm_tagme_mentions:
        
        
        if mention[3] > link_prob and mention[4] > rho:
            
            resource_label = mention[6].replace(' ', '_')
            resource = base_dbp + resource_label

            cm_mentions.append({
                'resource': resource,
                'ref': base_ref+str(ref_index),
                'spot': mention[0],
                'start': mention[1],
                'end': mention[2],
                'link_prob': mention[3],
                'rho': mention[4],
                'journal_id': mention[7].split('_')[0],
                'page_id': mention[7].replace('_', '-')
            })
            i += 1
            ref_index += 1
            
    print("Generated data for {} entity mentions from CM-tagme.".format(i))
    
    return cm_mentions
```

```python
# load tagme output (1.62 GB, may take a while)
cm_mentions = pickle.load(open("/data/cm/output/linker/cm_entities_tagme.pickle", 'rb'))
```

```python
# generate data
res_ref_data = generate_res_ref_data(cm_mentions, 0.6, 0.4)
```

```python
for r in res_ref_data[:10]:
    print(r)
```

```python
with open('/data/cm/output/generators/CM_Dataset/cm_tagme_resource_reference_data_06_04.pickle', 'wb') as outfile:
    pickle.dump(res_ref_data, outfile)
```
