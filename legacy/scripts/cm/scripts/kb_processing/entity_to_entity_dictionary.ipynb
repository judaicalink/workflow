```python
# goal is to create an entity-to-entity dictionary (graph) from wikipedia
# we reduce each wikipedia page to the set of entities that appear in that page
# based on that, for each entity we create a graph that represents its 'semantic space'
# the script takes approx 50 min to execute
import os, pprint, sys, time, pickle, json
from bs4 import BeautifulSoup as bs
root = "/data/cm/"
root_data = root+"data/"
```

```python
e2i = pickle.load(open(root+'data/wiki/de/e2i.pickle', 'rb'))
```

```python
# process wikipedia and create sets (bags) of entities for each document (page)
start_task = time.time()
c = 1
for folder in os.listdir(root_data+"wiki/de/unpacked/"):
    start = time.time()
    all_entities = []
    print("Processing ", folder, c, end=" ")
    
    for file in os.listdir(root_data+"wiki/de/unpacked/"+folder+"/"):
    
        filepath = root_data+"wiki/de/unpacked/"+folder+"/"+file
    
        input_file = open(filepath, "r", encoding="utf-8")
        data = input_file.read()
        soup = bs(data)
        docs = soup.findAll('doc')
        
        for doc in docs:
            
            try:
                title = doc['title']
                links = [a for a in doc.findAll('a') if a.has_attr('href')]
            
                if not title.startswith('Liste') and not title.endswith('Begriffskl√§rung)'):
                
                    doc_entities = set([e2i[link['href']] for link in links if not link['href'].startswith('http')])
            
                    all_entities.append(list(doc_entities))
            except KeyError:
                pass
    
    
    
    with open(root+'output/wiki/entity_to_entity/partials/'+folder+'.pickle', 'wb') as outfile:
        pickle.dump(all_entities, outfile)
        
    end = time.time()
    print(str(round(end-start, 2)), len(all_entities))
    c += 1
end_task = time.time()
print("Task executed in", str(round(end_task-start_task, 2)))
```
