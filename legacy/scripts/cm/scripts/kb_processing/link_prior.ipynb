```python
# Link prior is a measure of popularity for an entity and is defined as 
# "the fraction of all links in the knowledge repository that are incoming links to the given entity"
# a: number of links in the repository pointing to the entity
# b: overall number of links in the repository
# link prior = a/b
# output: {e: link_prior_score}

import os, pprint, pickle
root = "/data/cm/"
root_data = root+"data/"
```

```python
e2sf = pickle.load(open(root+'data/wiki/de/entity_to_sfs_count.pickle', 'rb'))
tot_links_wiki = 62967298 # total number of linked mentions extracted from dump
```

```python
l_prior = {}
for e, sfs in e2sf.items():
    a = sum(sfs.values())
    
    e_prior = a/tot_links_wiki
    l_prior[e] = e_prior
```

```python
with open(root+'data/wiki/de/link_prior.pickle', 'wb') as outfile:
    pickle.dump(l_prior, outfile)
```
