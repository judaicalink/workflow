```python
# scrape all metadata for compact memory (journal, issue, volume)
import os, pprint, requests, urllib, time, json
from bs4 import BeautifulSoup as bs
root = "/data/cm/data/xml/"
```

```python
# collect VLIDs by type (journal, issue, volume)
vlids = {}
journals, issues, volumes = [], [], []

for journal in os.listdir(root):
    journals.append(journal)
    if journal not in vlids.keys():
        vlids[journal] = {}
        
    if os.path.isdir(root+journal+"/"):        
        for volume in os.listdir(root+journal+"/"):
            volumes.append(volume)
            if volume not in vlids[journal].keys():
                vlids[journal][volume] = []
            
            if os.path.isdir(root+"/"+journal+"/"+volume+"/"):
                for issue in os.listdir(root+"/"+journal+"/"+volume+"/"):
                    issues.append(issue)
                    vlids[journal][volume].append(issue)
```

```python
j_metadata = json.load(open('j_metadata.json', 'r', encoding="utf-8"))

c = 0
for journal in journals:
    #c += 1
    #print(c, end="\t")
    if journal not in j_metadata.keys():
        print(journal)
        
        metadata_xml = requests.get('http://sammlungen.ub.uni-frankfurt.de/cm/oai?verb=GetRecord&metadataPrefix=fullmets&identifier=oai:sammlungen.ub.uni-frankfurt.de/cm:'+journal+'&mode=download')
        soup = bs(metadata_xml.text, 'xml')
        
        try:
            date_issued = soup.find('mods:dateIssued').string
        except AttributeError:
            date_issued = ""
            
        try:
            place = soup.find('mods:placeTerm').string
        except AttributeError:
            place = ""
            
        try:
            ident = soup.find('mods:identifier').string
        except AttributeError:
            ident = ""
            
        try:
            lang = soup.find('mods:language').string
        except AttributeError:
            lang = ""
            
        try:
            nonsort = soup.find('mods:titleInfo').find('mods:nonSort').string
        except AttributeError:
            nonsort = ""
            
        try:
            title = soup.find('mods:titleInfo').find('mods:title').string
        except AttributeError:
            title = ""
            
        try:
            subtitle = soup.find('mods:titleInfo').find('mods:subTitle').string
        except AttributeError:
            subtitle = ""
            
        try:
            publisher = soup.find('mods:publisher').string
        except AttributeError:
            publisher = ""
                
        full_title = " ".join([nonsort, title]).strip()
    
        j_metadata[journal] = {}
        
        j_metadata[journal]['dateIssued'] = date_issued
        j_metadata[journal]['place'] = place
        j_metadata[journal]['ident'] = ident
        j_metadata[journal]['lang'] = lang
        j_metadata[journal]['subtitle'] = subtitle
        j_metadata[journal]['title'] = full_title
        j_metadata[journal]['publisher'] = publisher
        
        with open('j_metadata.json', 'w', encoding='utf-8') as outfile:
            json.dump(j_metadata, outfile, indent=2, ensure_ascii=False)
        
        print(date_issued)
        print(place)
        print(ident)
        print(lang)
        print(full_title)
        print(subtitle)
        print(publisher)  
        print()
        
        date_issued, place, ident, lang, full_title, subtitle, publisher = "", "", "", "", "", "", ""
        
        time.sleep(5)
```
