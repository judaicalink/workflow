```python
# using partials surface form dictionaries w cout, build a global dictionary

import os, pprint, sys, time, pickle, json
from bs4 import BeautifulSoup as bs
root = "/data/cm/"
root_data = root+"data/"
```

```python
e2sf = {}
c = 1
for file in os.listdir(root+'output/wiki/surface_form_dictionary/partials/'):
    start = time.time()
    print(file, c, end=" ")
    
    with open(root+'output/wiki/surface_form_dictionary_count/partials/'+file, 'r', encoding='utf-8') as infile:
        
        partial = json.load(infile)
    
        for entity, sfs in partial.items():
            
            if not entity in e2sf:
                e2sf[entity] = sfs
            else:
                for sf, count in sfs.items():
                    if sf not in e2sf[entity]:
                        e2sf[entity][sf] = count
                    else:
                        e2sf[entity][sf] += count
    end= time.time()
    print('\t\t', str(round(end-start, 2)))
    c += 1
```

```python
with open(root+'/data/wiki/de/entity_to_sfs_count.pickle', 'wb') as outfile:
    pickle.dump(e2sf, outfile)
```

```python
sf2e = {}
for e, sfs in e2sf.items():
    for sf, count in sfs.items():
        if not sf in sf2e:
            sf2e[sf] = {e: count}
        else:
            if e not in sf2e[sf]:
                sf2e[sf][e] = count
            else:
                sf2e[sf][e] += count
```

```python
with open(root+'/data/wiki/de/sf_to_entities_count.pickle', 'wb') as outfile:
    pickle.dump(sf2e, outfile)
```
