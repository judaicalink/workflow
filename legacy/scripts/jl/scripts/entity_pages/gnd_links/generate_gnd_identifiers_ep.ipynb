```python
# for each entity page, generate the corresponding gnd ID
# (for pubby to be showed)
import os, pickle, pprint, json
```

```python
ep_index = pickle.load(open('/home/rovera/jl/scripts/entity_pages/ep/runs/run_0/ep_index.pickle', 'rb'))
```

```python
ep_gnd = {}
for ep, resources in ep_index.items():
    for resource in resources:
        if resource.startswith('http://d-nb.info/gnd/') and not resource.endswith('about'):
            gnd_id = resource.lstrip('http://d-nb.info/gnd/')
            if ep not in ep_gnd:
                ep_gnd[ep] = set([])
            ep_gnd[ep].add(gnd_id)
        elif resource.startswith('http://d-nb.info/gnd/') and resource.endswith('about'):
            gnd_id = resource.lstrip('http://d-nb.info/gnd/').rstrip('/about')
            if ep not in ep_gnd:
                ep_gnd[ep] = set([])
            ep_gnd[ep].add(gnd_id)
```

```python
c = 0
for k,v in ep_gnd.items():
    if len(v) > 1:
        c += 1
        print(k,v)
print(c)
```

```python
ep_gnd = {k: list(v)[0] for k,v in ep_gnd.items() if len(v) == 1}
```

```python
json.dump(ep_gnd, open('ep_GND_ids.json', 'w', encoding='utf-8'), indent=2)
```

```python
import os
from SPARQLWrapper import SPARQLWrapper, JSON
from rdflib import Graph, URIRef, Namespace, Literal
from rdflib.namespace import SKOS
```

```python
# query preferred labels
sparql = SPARQLWrapper("http://data.judaicalink.org/sparql/query")
sparql.setQuery("""
    PREFIX skos: <http://www.w3.org/2004/02/skos/core#>
    PREFIX owl: <http://www.w3.org/2002/07/owl#>
    PREFIX gnd: <http://d-nb.info/gnd/>
    PREFIX gndo: <http://d-nb.info/standards/elementset/gnd#>
    PREFIX dnbt: <http://d-nb.info/standards/elementset/dnb#>
    SELECT ?ep ?gndid
    WHERE {
        GRAPH <http://data.judaicalink.org/data/ep> {
            ?ep owl:sameAs ?o 
        }
        ?o gndo:gndIdentifier ?gndid 
        }
""")
sparql.setReturnFormat(JSON)
results = sparql.query().convert()
print(len(results['results']['bindings']))
```

```python
ep_gnd = {}
for res in results['results']['bindings']:

    ep = res['ep']['value']
    gndi = res['gndid']['value']
    if ep not in ep_gnd:
        ep_gnd[ep] = set([])
    ep_gnd[ep].add(gndi)
print(len(ep_gnd))
```

```python
for k,v in ep_gnd.items():
    if len(v) > 2:
        print(k)
        print(v)
        print()
```
