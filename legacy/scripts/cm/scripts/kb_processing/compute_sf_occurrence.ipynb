```python
# compute absolute surface form occurrence from raw documents

import os, pprint, sys, time, pickle, json
from bs4 import BeautifulSoup as bs
root = "/data/cm/"
root_data = root+"data/"
```

```python
sfs = pickle.load(open(root+'data/wiki/de/sf_to_entities_count.pickle', 'rb'))

sfs_count = {}
for sf, entities in sfs.items():
    tot = sum(entities.values())
    sfs_count[sf] = tot

print(len(sfs), len(sfs_count))
```

```python
print("Total number of linked mentions (links):", sum(sfs_count.values())) # useful for keyphraseness
```

```python
entities = pickle.load(open(root+'data/wiki/de/entity_to_sfs_count.pickle', 'rb'))

e_count = {}
for entity, sfs in entities.items():
    tot = sum(sfs.values())
    e_count[entity] = tot

print(len(entities), len(e_count))
```

```python
with open(root+'data/wiki/de/sf_count.pickle', 'wb') as outfile:
    pickle.dump(sfs_count, outfile)

with open(root+'data/wiki/de/e_count.pickle', 'wb') as outfile:
    pickle.dump(e_count, outfile)
```
