```python
# recognize named-entities and count their occurrence
import os, json, pprint, spacy
import langdetect
import time
path_data = "/data/cm/data/"
path_out = "/data/cm/output/ner/"
```

```python
nlp = spacy.load("de_core_news_lg")
```

```python
start = time.time()
entities = []
c = 0
for j in os.listdir(path_data+"pages/"):
    
    for page in os.listdir(path_data+"pages/"+j+"/"):
        
        if c % 100 == 0:
            print()
            print(c, end=" "*(5-len(str(c))))
        
        
        
        text = open(path_data+"pages/"+j+"/"+page, "r", encoding="utf-8").read()
        
        lang = None
        try:
            lang = langdetect.detect(text)
        except Exception:
            pass
        
        if lang == 'de':
        
        
            doc = nlp(text)
        
            for ent in doc.ents:
                e = {'text': ent.text, 'start': ent.start_char, 'end': ent.end_char, 'lab': ent.label_, 'page': page}
                entities.append(e)
            print("|", end="")
        else:
            print('/', end="")
        c += 1
end = time.time()
print("Done in {} s.".format(str(end-start)))
```

```python
# save result
with open(path_out+'entities_test.json', 'w', encoding='utf-8') as outfile:
    json.dump(entities, outfile, ensure_ascii=False, indent=2)
```

```python
pprint.pprint(entities[:10])
```

```python
# count entity-label couples, ex WIEN-LOC
e_lab_count = {}
for e in entities:
    e_lab = e['text'] + '__' + e['lab']
    if e_lab in e_lab_count.keys():
        e_lab_count[e_lab] += 1
    else:
        e_lab_count[e_lab] = 1
```

```python
e_sort = sorted([(k,v) for k,v in e_lab_count.items() if v > 1 and len(k) > 1], key=lambda x:x[1], reverse=True)
```
