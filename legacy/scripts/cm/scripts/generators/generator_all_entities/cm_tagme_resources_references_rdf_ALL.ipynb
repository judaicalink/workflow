Taking as input the data generated by the script ```cm_tagme_resource_reference_data```, this script generates the corresponding RDF graphs.

````python
import os, pickle, json, pprint
from rdflib import Graph, URIRef, Namespace, Literal, XSD
from rdflib.namespace import RDF, OWL
````

````python
with open('/data/cm/output/generators/CM_Dataset/cm_tagme_resource_reference_data_06_04.pickle', 'rb') as infile:
    data = pickle.load(infile)
````

````python
def instantiate_graph():
    cm_references = Graph()
    jl = Namespace("http://data.judaicalink.org/ontology/")
    cm_references.bind('rdf', RDF)
    cm_references.bind('jl', jl)
    return cm_references
````

````python
# instantiate graphs
cm_resources = Graph()
cm_references = Graph()
````

````python
# define Namespaces
jl = Namespace("http://data.judaicalink.org/ontology/")
````

````python
# bindings
cm_resources.bind('rdf', RDF)
cm_references.bind('rdf', RDF)
cm_resources.bind('jl', jl)
cm_references.bind('jl', jl)
````

````python
# may take a while...
mention_count = 0
chunks = 1
for mention in data:
    
    mention_count += 1
    resource = URIRef(mention['resource']) 
    reference = URIRef(mention['ref'])
    spot = Literal(mention['spot'])
    page = URIRef("http://data.judaicalink.org/data/compact-memory/"+mention['page_id'])
    start = Literal(mention['start'])
    end = Literal(mention['end'])
    link_prob = Literal(mention['link_prob'], datatype=XSD.float)
    rho = Literal(mention['rho'], datatype=XSD.float)
    # populate datasets    
    cm_resources.add((resource, jl.hasReference, reference))

    cm_references.add((reference, RDF.type, jl.Reference))
    cm_references.add((reference, jl.isOnPage, page))
    cm_references.add((reference, jl.hasSpot, spot))
    cm_references.add((reference, jl.hasStart, start))
    cm_references.add((reference, jl.hasEnd, end))
    cm_references.add((reference, jl.hasLinkProb, link_prob))
    cm_references.add((reference, jl.hasRho, rho))
    
    if mention_count % 100000 == 0:
        print("Saving chunk {}...".format(chunks), end='')
        cm_references.serialize(destination="/data/cm/output/generators/CM_Dataset/cm_references"+str(chunks)+".ttl", format="ttl")
        chunks += 1
        cm_references = instantiate_graph()
        print("Done!")
cm_references.serialize(destination="/data/cm/output/generators/CM_Dataset/cm_references"+str(chunks)+".ttl", format="ttl")
````

````python
print("Processed {} mentions.".format(mention_count))
print("Generated {} triples for resources.".format(len(cm_resources)))
#print("Generated {} triples for references.".format(len(cm_references)))
````
