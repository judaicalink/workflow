```python
import os, json, pprint, csv
from sklearn.cluster import KMeans
from sklearn import metrics
import numpy as np
path_data = "/data/cm/"
path = "/home/rovera/cm/"
```

```python
word_distances = list(csv.reader(open(path_data+"data/ocr_correction/word_distance.csv", 'r'), delimiter="\t"))
```

```python
X = np.array(word_distances)
X = X[:,1:]
X = np.array(X, dtype="float32")
X.shape
print(X[0])
```

```python
from sklearn.cluster import KMeans

previous_cost = 41292.40625

for k in range(2,3):

    print("Training KMeans with k ==", k, "...")
    # Create a kmeans model on our data, using k clusters.  random_state helps ensure that the algorithm returns the same results each time.
    kmeans_model = KMeans(n_clusters=k, random_state=1).fit(X)
    
    # These are our fitted labels for clusters -- the first cluster has label 0, and the second has label 1.
    labels = kmeans_model.labels_


    #create a dictionary to store clusters
    clu = {l: [] for l in labels}

    # populate cluster
    for i, l in enumerate(list(labels)):
        clu[l].append(word_distances[i][0])

    sil_avg = metrics.silhouette_score(X, kmeans_model.labels_)
    
    # Sum of distances of samples to their closest cluster center
    
    
    inertia = kmeans_model.inertia_
    print("k:",k, " cost:", inertia, 'gain:', previous_cost-inertia, ' silhouette avg:', sil_avg)
    for k,v in clu.items():
        print(k, len(v), v[100:110])
    print("Done!")
    print()
    
    previous_cost = inertia
```
