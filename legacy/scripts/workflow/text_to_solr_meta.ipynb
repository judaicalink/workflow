```python
# In this script we take as input the textual files for each page in compact memory
# we add the metadata about journal and volume for each page
# and then save this output in a json-like format that can be indexed in Solr
# !!! we only consider those journals that follow the canonical hierarchy journal --> volume --> issue
# which means that the final ES index will not contain all other journals

import os, json, pprint
root_data = '/data/cm/'
root = '/data/cm/scripts/'
```

```python
j_meta = json.load(open('/data/scripts/cm/scripts/metadata/j_metadata.json', 'r', encoding='utf-8'))
v_meta = json.load(open('/data/scripts/cm/scripts/metadata/v_metadata.json', 'r', encoding='utf-8'))
j_fields = {'dateIssued': '', 'place': '', 'ident': '', 'lang': '', 'subtitle': '', 'title': '', 'publisher': ''}
v_fields = {'dateIssuedVol': '', 'numberVol': ''}
```

```python
for journal in os.listdir(root_data+'output/text/'):
    
    # create a new folder for output (don't create the whole hierarchy, only journal folder)
    if not os.path.exists(root_data+"output/solr_meta/"+journal):
        os.makedirs(root_data+"output/solr_meta/"+journal)
        
    
    print('Processing', journal, '...')
    j_metadata = j_fields
    try:
        j_metadata = j_meta[journal]
    except KeyError as ke:
        pass

    if os.path.isdir(root_data+'output/text/'+journal+'/'):
        for volume in os.listdir(root_data+'output/text/'+journal+'/'):
            
            if os.path.isdir(root_data+'output/text/'+journal+'/'+volume+'/'):
                v_metadata = v_fields
                try:
                    v_metadata = v_meta[volume]
                except KeyError as ke:
                    pass
            
                for issue in os.listdir(root_data+'output/text/'+journal+'/'+volume+'/'):
                    
                    if os.path.isdir(root_data+'output/text/'+journal+'/'+volume+'/'+issue):
                
                        for page in os.listdir(root_data+'output/text/'+journal+'/'+volume+'/'+issue):
                        
                            if os.path.isfile(root_data+'output/text/'+journal+'/'+volume+'/'+issue+'/'+page):
                                # build the json file for SOLR
                            
                                filename = journal+'_'+volume+'_'+issue+'__'+page.split('.')[0]
                                text = open(root_data+'output/text/'+journal+'/'+volume+'/'+issue+'/'+page).read()
                                solr = []

                                # Solr document body
                                solr_doc = {
                                    "id": filename,
                                    "text": text,
                                    "dateIssued": j_metadata.get("dateIssued"),
                                    "lang": j_metadata.get("lang"),
                                    "place": j_metadata.get("place"),
                                    "ident": j_metadata.get("ident"),
                                    "subtitle": j_metadata.get("subtitle"),
                                    "title": j_metadata.get("title"),
                                    "publisher": j_metadata.get("publisher"),
                                    "dateIssuedVol": v_metadata.get("dateIssuedVol"),
                                    "numberVol": v_metadata.get("numberVol"),
                                }
                                
                                # JSONL line for Solr
                                line = json.dumps(solr_doc, ensure_ascii=False)
                                
                                # Append to output list
                                solr.append(line)
                                
                                with open(root_data+"output/solr_meta/"+journal+"/"+filename+".jsonl", "w", encoding="utf-8") as write_file:
                                    write_file.write("\n".join(solr))
                        
```
