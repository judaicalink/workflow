```python
import pprint, pickle, os, logging
from SPARQLWrapper import SPARQLWrapper, JSON
#from rdflib import Graph, URIRef
#from rdflib.namespace import RDF, OWL, RDFS
```

```python
sparql = SPARQLWrapper("http://de.dbpedia.org/sparql")
sparql.setQuery("""
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX dbo: <http://dbpedia.org/ontology/>
    SELECT ?s
    WHERE {?s rdf:type dbo:Location}
""")
sparql.setReturnFormat(JSON)
results = sparql.query().convert()
```

```python
# get all places
locations = set([])
for res in results['results']['bindings']:
    location = res['s']['value']
    locations.add(location)
print("Found {} locations in dbpedia-DE.".format(len(locations)))
```

```python
sparql = SPARQLWrapper("http://de.dbpedia.org/sparql")
sparql.setQuery("""
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX dbo: <http://dbpedia.org/ontology/>
    SELECT ?s
    WHERE {?s rdf:type dbo:Person}
""")
sparql.setReturnFormat(JSON)
results = sparql.query().convert()
```

```python
# get all places
persons = set([])
for res in results['results']['bindings']:
    person = res['s']['value']
    persons.add(person)
print("Found {} locations in dbpedia-DE.".format(len(persons)))
```
