```python
all_mentions = {}
for file in os.listdir(root+'output/wiki/'):
    
    data = json.load(open(root+'output/wiki/'+file, 'r', encoding='utf-8'))
    print('Processing', file, end=' ')
    
    for k,v in data.items():
        if k in all_mentions.keys():
            all_mentions[k].update(v)
        else:
            all_mentions[k] = set(v)
    print('\t\t', sys.getsizeof(all_mentions)/1000000)
```

```python
# entity to surface forms dictionary
entity_to_sfs = {k: list(v) for k,v in all_mentions.items()}
```

```python
# surface forms to entity dictionary
sf_to_entities = {}
for entity, sfs in entity_to_sfs.items():
    for sf in sfs:
        if sf not in sf_to_entities.keys():
            sf_to_entities[sf] = set()
        sf_to_entities[sf].add(entity)
sf_to_entities = {k: list(v) for k,v in sf_to_entities.items()}
```

```python
with open(root+'data/wiki/de/entity_to_sfs.json', 'w', encoding='utf-8') as outfile:
    json.dump(entity_to_sfs, outfile, indent=2)
    
with open(root+'data/wiki/de/sf_to_entities.json', 'w', encoding='utf-8') as outfile:
    json.dump(sf_to_entities, outfile, indent=2)
```
