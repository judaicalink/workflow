```python
# compute simple word count for the whole CM 
import os, pprint, json, pickle
path_data = "/data/cm/"
path = "/home/rovera/cm/"
```

```python
voc = {}
words = 0
tokens = 0
langs = {}

for j in os.listdir(path_data+'output/nlp_pipeline/tokenized_CM/'):
    
    print("Processing {}".format(j))
    
    for file in os.listdir(path_data+'output/nlp_pipeline/tokenized_CM/'+j+'/'):
        
        page = json.load(open(path_data+'output/nlp_pipeline/tokenized_CM/'+j+'/'+file, 'r', encoding="utf-8"))
        
        words += page['wc']
        tokens += page['tc']
        
        lang = page['lang']
        if lang in langs.keys():
            langs[lang] += 1
        else:
            langs[lang] = 1
            
        for token in page['tokens']:
            if token in voc:
                voc[token] += 1
            else:
                voc[token] = 1
```

```python
print("All words:", words)
print("All tokens:", tokens)
print("Vocabulary size:", len(voc))
print("Tokens with more than 10 occurs:", len({k:v for k,v in voc.items() if v > 5}))
for lang, count in langs.items():
    print(lang, count/sum(langs.values())*100)
```

```python
with open(path_data+'output/nlp_pipeline/word_count_CM/word_count_CM.pickle', 'wb') as outfile:
    pickle.dump(voc, outfile)
```
