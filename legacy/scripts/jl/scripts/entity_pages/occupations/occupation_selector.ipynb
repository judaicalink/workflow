```python
import pprint, os
from SPARQLWrapper import SPARQLWrapper, JSON
from rdflib import Graph, URIRef, Namespace, Literal
```

```python
sparql = SPARQLWrapper("http://data.judaicalink.org/sparql/query")
sparql.setQuery("""
    PREFIX jl: <http://data.judaicalink.org/ontology/>
    PREFIX owl: <http://www.w3.org/2002/07/owl#>
    SELECT ?ep ?occupation
    WHERE {
        GRAPH <http://data.judaicalink.org/data/ep> {
            ?ep owl:sameAs ?same 
        }
        ?same jl:occupation ?occupation
        }
""")
sparql.setReturnFormat(JSON)
results = sparql.query().convert()
```

```python
print(len(results['results']['bindings']))
```

```python
occupations = {}
for res in results['results']['bindings']:
    ep = res['ep']['value']
    occ = res['occupation']['value'].split('/')[-1]
    if ep not in occupations:
        occupations[ep] = {}
    if occ not in occupations[ep]:
        occupations[ep][occ] = 0
    occupations[ep][occ] += 1
```

```python
for k,v in occupations.items():
    if len(v) > 1:
        print(k,v)
```
