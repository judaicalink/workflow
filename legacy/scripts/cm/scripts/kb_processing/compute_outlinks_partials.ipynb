```python
import os, pprint, sys, time, pickle, json
from bs4 import BeautifulSoup as bs
from urllib.parse import unquote
root = "/data/cm/"
root_data = root+"data/"
```

```python
start_task = time.time()
c = 1
for folder in os.listdir(root_data+"wiki/de/unpacked/"):
    start = time.time()
    all_outlinks = {}
    print("Processing ", folder, c, end=" ")
    
    for file in os.listdir(root_data+"wiki/de/unpacked/"+folder+"/"):
    
        filepath = root_data+"wiki/de/unpacked/"+folder+"/"+file
    
        input_file = open(filepath, "r", encoding="utf-8")
        data = input_file.read()
        soup = bs(data)
        docs = soup.findAll('doc')
        
        for doc in docs:
            
            try:
                title = doc['title']
                if title not in all_outlinks:
                    all_outlinks[title] = {}
                    
                links = [a for a in doc.findAll('a') if a.has_attr('href')]
            
                if not title.startswith('Liste') and not title.endswith('Begriffskl√§rung)'):
                
                    
                    doc_outlinks = set([unquote(link['href']) for link in links if not link['href'].startswith('http')])
            
                    for outlink in doc_outlinks:  #update global dict with found outlinks
                        if not outlink in all_outlinks[title]:
                            all_outlinks[title][outlink] = 0
                        all_outlinks[title][outlink] += 1
                        
                        
            except KeyError:
                pass
    
    
    
    with open(root+'output/wiki/outlinks/partials/'+folder+'.pickle', 'wb') as outfile:
        pickle.dump(all_outlinks, outfile)

    end = time.time()
    print(str(round(end-start, 2)), len(all_outlinks), len(all_outlinks.values()))
    c += 1
end_task = time.time()
print("Task executed in", str(round(end_task-start_task, 2)))
```
