```python
# using partials surface form dictionaries, build a global dictionary

import os, pprint, sys, time, pickle, json
from bs4 import BeautifulSoup as bs
root = "/data/cm/"
root_data = root+"data/"
```

```python
e2sf = {}

for file in os.listdir(root+'output/wiki/surface_form_dictionary/partials/'):
    print(file)
    
    with open(root+'output/wiki/surface_form_dictionary/partials/'+file, 'r', encoding='utf-8') as infile:
        
        partial = json.load(infile)
    
        for entity, sfs in partial.items():
            
            if entity in e2sf:
                e2sf[entity].update(sfs)
            else:
                e2sf[entity] = set(sfs)
        
```

```python
with open(root+'/data/wiki/de/entity_to_sfs.pickle', 'wb') as outfile:
    pickle.dump(e2sf, outfile)
```

```python
sf2e = {}
for e, sfs in e2sf.items():
    for sf in sfs:
        if not sf in sf2e:
            sf2e[sf] = {e}
        else:
            sf2e[sf].add(e)
```

```python
with open(root+'/data/wiki/de/sf_to_entities.pickle', 'wb') as outfile:
    pickle.dump(sf2e, outfile)
```
