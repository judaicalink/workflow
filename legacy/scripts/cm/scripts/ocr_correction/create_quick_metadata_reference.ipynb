```python
# create a quick reference metadata for each vlid
# {'vlid': 'y': year, 'f': filename, 'j_id': journal_id, 'jn': journal_name, 'pdf': if a pdf is available on VL Fra, True/False}

import os, json, pprint, csv
path_data = "/data/cm/"
path = "/home/rovera/cm/"
```

```python
metadata = list(csv.reader(open(path_data+"data/metadata/CM_Seiten_Metadaten.csv"), delimiter='\t'))
```

```python
# these journals do not have access as pdf of visual library
blacklist = ('2431292', '2823768', '10112841', '4086896', '9038025', '4875667', '7938572', '8553624', '8823924', '9498581', '9572329', '9616703', '9620162')
```

```python
vlid_to_meta = {line[0]: {'y': line[10].split('-')[0], 'f': "", 'j_id': line[2], 'jn': line[6], 'pdf': None} for line in metadata[1:]}
```

```python
for vlid, val in vlid_to_meta.items():
    if vlid_to_meta[vlid]['j_id'] in blacklist:
        vlid_to_meta[vlid]['pdf'] = False
    else:
        vlid_to_meta[vlid]['pdf'] = True
```

```python
for journal in os.listdir(path_data+"output/nlp_pipeline/tokenized_CM/"):    
        
        for page in os.listdir(path_data+"output/nlp_pipeline/tokenized_CM/"+journal+"/"):
            
            if page.endswith('.json'):
                
                vlid = page.strip('.json').split('_')[-1]
                #print(vlid, end=" ")
                
                try:
                    vlid_to_meta[vlid]['f'] = page.strip('.json')
                    
                except KeyError:
                    pass
```

```python
vlid_to_meta['3981046']
```

```python
with open(path_data+"data/metadata/short_metadata/vlid_to_meta.json", 'w', encoding="utf-8") as outfile:
    json.dump(vlid_to_meta, outfile, indent=2, ensure_ascii=False)
```

```python
# inverse dictionary 'page_name': {metadata}
pagename_to_meta = {}
for vlid, val in vlid_to_meta.items():
    
    pagename = vlid_to_meta[vlid]['f']
    pagename_to_meta[pagename] = {k:v for k,v in val.items() if k != 'f'}
    pagename_to_meta[pagename]['vlid'] = vlid
```

```python
with open(path_data+"data/metadata/short_metadata/filename_to_meta.json", 'w', encoding="utf-8") as outfile:
    json.dump(pagename_to_meta, outfile, indent=2, ensure_ascii=False)
```
