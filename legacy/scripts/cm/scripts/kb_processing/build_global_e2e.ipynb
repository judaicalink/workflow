```python
# using the partials, build a global entity to entity dictionary where for each entity the absolute (unique) frequency
# with other entities is computed
import os, json, pprint, pickle, os, time
root = "/data/cm/"
root_data = root+"data/"
```

```python

e2e = {}

c = 1
for part in os.listdir(root+'output/wiki/entity_to_entity/partials_dict/'):
    
    start = time.time()
    print(part, end=' ')
        
    with open(root+'output/wiki/entity_to_entity/partials_dict/'+part, 'rb') as file:
        
        partial = pickle.load(file)
        
        for k, v in partial.items():
            if k not in e2e:
                e2e[k] = v
            else:
                for e, count in v.items():
                    if count > 0:
                        if not e in e2e[k]:
                            e2e[k][e] = count
                        else:
                            e2e[k][e] += count
    end = time.time()
    print(str(round(end-start, 0)), c)
    c += 1
e2e = {k:v for k,v in e2e.items() if len(v) > 0}
print(len(e2e))
pickle.dump(e2e, open(root+'output/wiki/entity_to_entity/e2e.pickle', 'wb'))
```
