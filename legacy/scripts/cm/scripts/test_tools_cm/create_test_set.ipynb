```python
# create a test set of 100 documents for testing performance of different tools like TagMe, Spacy NER, etc
import os, json, pprint
import random
from shutil import copyfile
path_data = "/data/cm/"
path = 'home/rovera/cm/'
```

```python
journal_status = json.load(open(path_data+"output/linker/indexes/journals_status.json", "r", encoding="utf-8"))
```

```python
processed = [j for j, val in journal_status.items() if val is True]
```

```python
test_samples = []

for j in processed:
    
    candidates = []
    
    for root, dirs, files in os.walk(path_data+"output/text/"+j+"/"):
        for file in files:
            fp = os.path.join(root, file)
            candidates.append(fp)
            
    random.shuffle(candidates)
    for c in candidates[:4]:
        test_samples.append(c)
```

```python
for ts in test_samples:
    
    src_root = '/data/cm/output/text/'
    dest_root = '/data/cm/output/test_tools_cm/test_data/'
    fpath = ts.split('.')[0].split('text/')[1]
    fname = fpath.replace('_', '__').replace('/', '_')
    
    copyfile(src_root+fpath+'.txt', dest_root+fname+'.txt')
    print(fname)
print("Done!")
```
