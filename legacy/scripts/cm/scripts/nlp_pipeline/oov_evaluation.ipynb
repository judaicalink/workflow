```python
import os, json, pprint
path_data = "/data/cm/"
path = "/home/rovera/cm/"
```

```python
oov_eval = json.load(open(path_data+"output/nlp_pipeline/oov_count/oov_words_eval_filled.json", "r", encoding="utf-8"))
```

```python
evaluation = {}
for k,v in oov_eval.items():
    if v['state'] != "":
        if v['state'] not in evaluation:
            evaluation[v['state']] = 0
        evaluation[v['state']] += 1
pprint.pprint(evaluation)
print(sum(evaluation.values()))
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = list(evaluation.values())
y = list(evaluation.keys())
x_perc = [el/sum(x) for el in x]
print(x_perc)

err_types = y
y_pos = np.arange(9)
counts = x_perc

plt.bar(y_pos, counts, align='center', alpha=1)
plt.xticks(y_pos, err_types)
plt.ylabel('%')
plt.title('Source out-of-vocabulary words in CM (N = 310)')

plt.show()
```
