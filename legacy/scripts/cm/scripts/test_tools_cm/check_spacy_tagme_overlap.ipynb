```python
# check overlapping annotations in the output of spacy and tagme
import os, json, pprint, pickle
path_data = "/data/cm/"
path = "/home/rovera/cm/"
```

```python
spacy_out = pickle.load(open(path_data+"output/test_tools_cm/spacy_linking/tagging_output.pickle", 'rb'))
tagme_out = pickle.load(open(path_data+"output/test_tools_cm/tagme_linking/tagging_output.pickle", 'rb'))
```

```python
ex = "4781937_4782138_237__4782177"
print(spacy_out[ex])
print(tagme_out[ex])
```

```python
# collect overlapping mentions
over = {}
for doc, val in tagme_out.items():
    
    over[doc] = set()
    
    for tagme_mention in val:
        curr_tagme_ment = tagme_mention['mention']
        
        for spacy_mention in spacy_out[doc]:
            if spacy_mention['mention'] == curr_tagme_ment:
                
                curr_overlap = []
                curr_overlap.append(tagme_mention['mention'])
                curr_overlap.append(tagme_mention['e_title'])
                curr_overlap.append(tagme_mention['e_id'])
                curr_overlap.append(tagme_mention['score'])
                curr_overlap.append(spacy_mention['label'])
                curr_overlap.append(tagme_mention['b'])
                
                
                over[doc].add(tuple(curr_overlap))
```

```python
# create file for evaluation
evaluation = {}
for k,v in over.items():
    
    evaluation[k] = {}
    
    for val in v:
        val = [str(el) for el in val]
        val_to_str = "__".join(val)
        evaluation[k][val_to_str] = -1
```

```python
pprint.pprint(evaluation)
```

```python
with open(path+"scripts/test_tools_cm/evaluation_set.json", "w", encoding="utf-8") as outfile:
    json.dump(evaluation, outfile, ensure_ascii=False, indent=2)
```

```python
spacy_mentions = []
tagme_mentions = []
spacy_unique = None
tagme_unique = None
docs = list(spacy_out.keys())

overlap_ratios = []

for doc in docs:
    spacy_mentions, tagme_mentions = [], []
    print(doc)
    
    for v in spacy_out[doc]:
    
        spacy_mentions.append(v['mention'])
    spacy_unique = set(spacy_mentions)
        
    for v in tagme_out[doc]:
        tagme_mentions.append(v['mention'])
    tagme_unique = set(tagme_mentions)
        
    print("Spacy all spots:", len(spacy_mentions))
    print("Spacy unique:", len(spacy_unique))
    print("Tagme all spots:", len(tagme_mentions))
    print("Tagme unique:", len(tagme_unique))
    print("Found {} unique entities overlap".format(len(spacy_unique & tagme_unique)))
    try:
        o_ratio = len(spacy_unique & tagme_unique)/len(spacy_unique)
    except DivisionByZeroError:
        o_ratio = 0.0
    print("Overlap ratio:", o_ratio)
    overlap_ratios.append(o_ratio)
    print()
```

```python
print("Average overlap ratio:", sum(overlap_ratios)/len(overlap_ratios))
```
