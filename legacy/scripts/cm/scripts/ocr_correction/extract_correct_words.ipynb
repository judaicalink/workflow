```python
import os, pprint, csv, json, gensim, random, pickle
path_data = "/data/cm/"
path = "/home/rovera/cm/"
# 1) extract the word length distribution of wrong word
# 2) extract the same word length distribution also from correct words
```

```python
wrong = pickle.load(open(path_data+"data/ocr_correction/training_data/wrong.pickle", 'rb'))
correct = pickle.load(open(path_data+"data/ocr_correction/training_data/correct.pickle", 'rb'))
```

```python
def count_word_len(pairs):
    word_len = {}
    for pair in pairs:
        if " " not in pair[1]:
            if len(pair[1]) not in word_len:
                word_len[len(pair[1])] = 0
            word_len[len(pair[1])] += 1
    return word_len
```

```python
word_len = count_word_len(wrong)
pprint.pprint(word_len)
```

```python
def extract_words_by_len(pairs, word_len_dict):
    random.shuffle(pairs)
    out_word_pairs = []
    for length, count in word_len_dict.items():
        counter = 0
        for i in range(len(pairs)):
            if len(pairs[i][0]) == length:
                out_word_pairs.append(pairs[i])
                counter += 1
            if counter == count:
                break
    return out_word_pairs
```

```python
correct_out = extract_words_by_len(correct, word_len)
```

```python
random.shuffle(correct_out)
pprint.pprint(correct_out[500:550])
```
