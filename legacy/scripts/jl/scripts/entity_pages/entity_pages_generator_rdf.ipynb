```python
import os, pickle, pprint
from rdflib import Graph, URIRef, Namespace
from rdflib.namespace import RDF, OWL
```

```python
ep_dict = pickle.load(open('/home/rovera/jl/output/entity_pages/ep_data.pickle', 'rb'))
```

```python
ep = Graph()
ep.bind('owl', OWL)
```

```python
for ep_uri, aliases in ep_dict.items():
    for alias in aliases:
        s = URIRef(ep_uri)
        o = URIRef(alias)
        ep.add((s, OWL.sameAs, o))
```

```python
ep.serialize(destination="/home/rovera/jl/output/entity_pages/entity_pages.ttl", format="ttl")
```
