```python
import os, json, csv, pprint, pickle
```

```python
def tagme2pickle(path_in, path_out, impatience_manager=True):
    """
    Create a csv dataset for the entities extracted by TagMe from Compact Memory
    Each line of the output has the following schema
    surface form | start | end | link probability | rho | wikipedia id | wikipedia title | path of the CM page
    """
    print("Collecting entities from TagMe output (could last a while)...")
    c = 0
    dataset = []
    for folder in os.listdir(path_in):
        
        for file in os.listdir(path_in+folder+'/'):
            page_matches = json.load(open(path_in+folder+'/'+file, 'r', encoding="utf-8"))
            
            for match in page_matches:
                entity = []
                page_path = file.split('.json')[0]
                entity.append(match['spot'])
                entity.append(match['start'])
                entity.append(match['end'])
                entity.append(match['link_probability'])
                entity.append(match['rho'])
                entity.append(match['id'])
                try:
                    entity.append(match['title'])
                    entity.append(page_path)
                    dataset.append(entity)
                except KeyError:
                    pass
                    #print(match)
        
        if impatience_manager is True:
            print("|", end="")
        

    print()            
    print("Collected {} entity mentions.".format(len(dataset)))
    print("Saving pickle to {}...".format(path_out), end=" ")
    with open(path_out+"cm_entities_tagme.pickle", 'wb') as outfile:
        pickle.dump(dataset, outfile)
        #csv_writer = csv.writer(csvfile, delimiter='\t', quotechar='"')
        #for datapoint in dataset:
        #    csv_writer.writerow(datapoint)
    print("Done!")
```

```python
path_in = '/data/cm/output/linker/entities/'
path_out = '/data/cm/output/linker/'
```

```python
tagme2pickle(path_in, path_out)
```
