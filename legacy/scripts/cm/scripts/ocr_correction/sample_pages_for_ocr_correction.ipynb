```python
# create a sample of CM for annotation of OCR errors
# the sample is flatly distributed over the time of publication
# 1000 documents per decade

import os, pprint, json, pprint, pickle
import random
import gensim
from langdetect import detect
path_data = "/data/cm/"
path = "/home/rovera/cm/"
```

```python
we = gensim.models.KeyedVectors.load_word2vec_format(path_data+'word_emb/cc.de.300.vec', binary=False)
```

```python
pages_by_year = json.load(open(path_data+"data/metadata/short_metadata/pages_by_year.json", 'r', encoding='utf-8'))
```

```python
l = [len(v) for v in pages_by_year.values()]
print(sum(l))
```

```python
sample = []
for k,v in pages_by_year.items():
    decade = v
    random.shuffle(decade)
    if len(decade) >= 3000:
        for el in decade[:3000]:
            sample.append(el)
    else:
        for el in decade:
            sample.append(el)
```

```python
random.shuffle(sample)
pprint.pprint(sample[:10])
print(len(sample))
```

```python
# extract words from tokenized
# without language detection

existing_tokens = json.load(open(path_data+"data/ocr_correction/empty/annotation_set_1.json", 'r', encoding="utf-8"))
existing_tokens = list(existing_tokens.keys())
    
annotation_set = {}
for filename in sample:
    
    journal = filename.split('_')[0]
    
    file = json.load(open(path_data+"output/nlp_pipeline/tokenized_CM/"+journal+"/"+filename+".json", 'r', encoding="utf-8"))
    
    for token in file['tokens']:
        
        if token not in existing_tokens:
          
            try:
                we[token]
        
            except KeyError: # the token is out-of-vocabulary
            
                if token not in annotation_set:
                    annotation_set[token] = {
                        'journal': journal,
                        'doc': filename,
                        'correct': "",
                        'index': token_i
                        }

        if len(annotation_set) % 1000 == 0:
            print(len(annotation_set), end=" ")
    if len(annotation_set) >= 20000:
        break
```

```python
# extract words from tokenized
# with language detection

existing_tokens = json.load(open(path_data+"data/ocr_correction/empty/annotation_set_1.json", 'r', encoding="utf-8"))
existing_tokens = list(existing_tokens.keys())
    

annotation_set = {}
for filename in sample:
    
    journal = filename.split('_')[0]
    
    file = json.load(open(path_data+"output/nlp_pipeline/tokenized_CM/"+journal+"/"+filename+".json", 'r', encoding="utf-8"))
    
    for token in file['tokens']:
        
        if token not in existing_tokens:
        
            try:
                lang = detect(token) # detect language
        
                if lang != 'he': # filter out hebrew words
        
                    token_i = file['tokens'].index(token)
        
                    try:
                        we[token]
        
                    except KeyError: # the token is out-of-vocabulary
            
                        if token not in annotation_set:
                            annotation_set[token] = {
                            'journal': journal,
                            'doc': filename,
                            'correct': "",
                            'index': token_i
                            }
            except Exception:
                #print(token, end=" ")
                pass
            
        #if len(annotation_set) % 1000 == 0 and len(annotation_set) != 0:
        #    print(len(annotation_set), end=" ")
    print(len(annotation_set), end=" ")
    if len(annotation_set) >= 20000:
        break
```

```python
i = 0
for k,v in annotation_set.items():
    if i > 3500 and i < 3580:
        print(k)
        print(v)
        print()
    i+= 1
```

```python
with open(path_data+"data/ocr_correction/empty/annotation_set_2.json", 'w', encoding="utf-8") as outfile:
    json.dump(annotation_set, outfile, indent=2, ensure_ascii=False)
```

```python
# create csv version
import csv
annotation_set_list = []
for k,v in annotation_set.items():
    item = []
    item.append(k)
    item.append("")
    item.append(v['journal'])
    item.append(v['doc'])
    
    item.append(v['index'])
    annotation_set_list.append(item)
print(len(annotation_set_list))
```

```python
with open(path_data+"data/ocr_correction/empty/annotation_set_2.csv", 'wt') as out_file:
    tsv_writer = csv.writer(out_file, delimiter='\t')
    for line in annotation_set_list:
        tsv_writer.writerow(line)
```
