```python
# produce ngrams from text
import os, json, pprint, nltk, pickle
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.util import ngrams
root = '/data/cm/'
root_data = root+'data/'
```

```python
sf_to_entities = pickle.load(open(root+'data/wiki/de/sf_to_entities.pickle', 'rb'))
```

```python
sf_to_entities['Die VÃ¶lker']
```

```python
# filtering of surface forms
surfaces = list(sf_to_entities.keys())
surfaces = [s for s in surfaces if s is not None]
surfaces = [s for s in surfaces if len(s) > 2 and s[0].isupper()]
```

```python
for journal in os.listdir(root+'data/pages/'):
    
    print(journal)
    journal_candidates = {}
    
    for file in os.listdir(root+'data/pages/'+journal+'/'):
    
        text = open(root+'data/pages/'+journal+'/'+file, 'r', encoding='utf-8').read()
        words = word_tokenize(text)
        
        unigrams = list(ngrams(words, 1))
        bigrams = list(ngrams(words, 2))
        trigrams = list(ngrams(words, 3))
        quadgrams = list(ngrams(words, 4))
        pentagrams = list(ngrams(words, 5))
        esagrams = list(ngrams(words, 6))
    
        n_grams = set()
        for gramclass in (unigrams, bigrams, trigrams, quadgrams, pentagrams, esagrams):
            for gram in gramclass:
                cand = " ".join(gram)
                if len(cand) > 2:
                    n_grams.add(cand)
                    
        candidates = set()
        for n_gram in n_grams: # filter ngrams with surface forms available in Wikipedia
            try:
                sf_to_entities[n_gram]
                candidates.add(n_gram)
            except KeyError:
                pass
        journal_candidates[file.split('.')[0]] = list(candidates)
        
    with open(root+'output/mention_detection/ngrams/'+journal+'/mentions.json', 'w', encoding='utf-8') as outfile:
        json.dump(journal_candidates, outfile, indent=2, ensure_ascii=False)
```
