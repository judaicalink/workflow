```python
# using the e2e dictionary, where the absolute frequency of cooccurrence is stored, compute pmi for each pair
# of entities. In order to do that, we need the total (unique) number of entities occurring in the wiki dump
import os, json, pprint, pickle, os, time, math
root = "/data/cm/"
root_data = root+"data/"
```

```python
unique_e = pickle.load(open(root+'data/wiki/de/unique_entity_count.pickle', 'rb'))
i2e = pickle.load(open(root+'data/wiki/de/i2e.pickle', 'rb'))
e2e = pickle.load(open(root+'output/wiki/entity_to_entity/e2e.pickle', 'rb'))
```

```python
tot = sum(unique_e.values())
```

```python
e2e_pmi = {}

c = 0
for e1, entities in e2e.items():
    if entities != {}:
    
        if not e1 in e2e_pmi:
            e2e_pmi[e1] = {}
        c += 1
        for e2, abs_cooc in entities.items():
            e1_name = i2e[e1]
            e2_name = i2e[e2]
            p_e1 = unique_e[e1_name]/tot
            p_e2 = unique_e[e2_name]/tot
            p_cooc_e1e2 = abs_cooc/tot
        
            pmi_e1e2 = math.log(p_cooc_e1e2/(p_e1*p_e2))
            e2e_pmi[e1][e2] = pmi_e1e2
```

```python
e2e_pmi = {k:v for k,v in e2e_pmi.items() if len(v) > 0}
pickle.dump(e2e_pmi, open(root+'output/wiki/entity_to_entity/e2e_pmi.pickle', 'wb'))
```
