```python
locs = tuple([
    "http://dbpedia.org/ontology/Place",
    "http://dbpedia.org/ontology/Location",
    "http://schema.org/Place"
])

pers = tuple([
    "http://xmlns.com/foaf/0.1/Person",
    "http://dbpedia.org/ontology/Person",
    "http://schema.org/Person"
])

orgs = tuple([
    "http://schema.org/Organization",
    "http://dbpedia.org/ontology/Organisation"
])
events = tuple([
    "http://schema.org/Event",
    "http://dbpedia.org/ontology/Event"
])

filtered_entities = {}

for entity, vals in entity_classes.items():
    if len(entity) > 2:
        for cat in vals['cats']:
            if cat in locs:
                filtered_entities[entity] = 'http://dbpedia.org/ontology/Location'
                break
            elif cat in orgs:
                filtered_entities[entity] = 'http://dbpedia.org/ontology/Organization'
                break
            elif cat in pers:
                filtered_entities[entity] = 'http://dbpedia.org/ontology/Person'
                break
            elif cat in events:
                filtered_entities[entity] = 'http://dbpedia.org/ontology/Event'
                break
```
