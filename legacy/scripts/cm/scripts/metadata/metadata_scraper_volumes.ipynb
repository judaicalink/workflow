```python
# scrape all metadata for compact memory (journal, issue, volume)
import os, pprint, requests, urllib, time, json
from bs4 import BeautifulSoup as bs
root = "/data/cm/data/xml/"
```

```python
# collect VLIDs by type (journal, issue, volume)
vlids = {}
journals, issues, volumes = [], [], []

for journal in os.listdir(root):
    journals.append(journal)
    if journal not in vlids.keys():
        vlids[journal] = {}
        
    if os.path.isdir(root+journal+"/"):
        for volume in os.listdir(root+journal+"/"):
            volumes.append(volume)
            if volume not in vlids[journal].keys():
                vlids[journal][volume] = []
            
            if os.path.isdir(root+"/"+journal+"/"+volume+"/"):
                for issue in os.listdir(root+"/"+journal+"/"+volume+"/"):
                    issues.append(issue)
                    vlids[journal][volume].append(issue)
```

```python
v_metadata = json.load(open('v_metadata.json', 'r', encoding="utf-8"))

c = 0
for volume in volumes:
    #c += 1
    #print(c, end="\t")
    if volume not in v_metadata.keys():
        print(volume)
        
        metadata_xml = requests.get('http://sammlungen.ub.uni-frankfurt.de/cm/oai?verb=GetRecord&metadataPrefix=fullmets&identifier=oai:sammlungen.ub.uni-frankfurt.de/cm:'+volume+'&mode=download')
        soup = bs(metadata_xml.text, 'xml')
        
        try:
            date_issued = soup.find('mods:dateIssued').string
        except AttributeError:
            date_issued = ""
            
        try:
            number = soup.find('mods:detail').find('mods:number').string
        except AttributeError:
            number = ""
            
        
    
        v_metadata[volume] = {}
        
        v_metadata[volume]['dateIssuedVol'] = date_issued
        v_metadata[volume]['numberVol'] = number
        
        
        with open('v_metadata.json', 'w', encoding='utf-8') as outfile:
            json.dump(v_metadata, outfile, indent=2, ensure_ascii=False)
        
        print(date_issued)
        print(number)
        print()
        
        date_issued, number = "", ""        
        time.sleep(4)
```
