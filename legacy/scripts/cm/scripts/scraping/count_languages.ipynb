```python
import os, pprint, json
root = 'C:/Users/rovera/compact_memory/'
```

```python
data = json.load(open(root+'data/journal_metadata/journal_metadata_title_lang.json', 'r', encoding="utf-8")) 
```

```python
langs = {}
for k,v in data.items():
    lang = v['lang'].lower()
    if lang not in langs.keys():
        langs[lang] = 0
    langs[lang] += 1
```

```python
pprint.pprint(langs)
```
