```python
# take as input che cooccurrences computed by the script name-extractor2
# compute cooccurrence + string similarity
# output a graph

import os, json, pprint
root = "C:/Users/rovera/compact_memory/"
```

```python
cooc = json.load(open(root+'data/gnd/cooc.json', "r", encoding="utf-8"))
```

```python
print(len(cooc['gndo:forename']))
```

```python
# dict with same structure as cooc, but with relative cooccurrence instead of absolute
rel_cooc = {}
rel_cooc['gndo:forename'] = {k: 0.0 for k in cooc['gndo:forename']}
```

```python
for target, cooccurrents in cooc['gndo:forename'].items():
    tot_occurrences = sum(cooccurrents.values())
    rel_cooc['gndo:forename'][target] = {}
    for cooccurrent, count in cooccurrents.items():
        relative_cooc = round(count/tot_occurrences, 4)
        rel_cooc['gndo:forename'][target][cooccurrent] = relative_cooc
```

```python
for k,v in rel_cooc['gndo:forename'].items():
    print(sum(v.values()))
    print(k,v)
    break
```
