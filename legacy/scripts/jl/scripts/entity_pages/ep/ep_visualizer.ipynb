```python
import os, pickle, pprint
ep_index = pickle.load(open('~/jl/scripts/entity_pages/version_5/runs/run_0/ep_index.pickle', 'rb'))
ep_inv_index = pickle.load(open('~/jl/scripts/entity_pages/version_5/runs/run_0/ep_inv_index.pickle', 'rb'))
```

```python
for k,v in ep_index.items():
    print(k)
    print(v)
    print()
    break
    
for k,v in ep_inv_index.items():
    print(k)
    print(v)
    print()
    break
```

```python
def query_ep(query, ep_index, ep_inv_index):
    print("Query: ", query)
    if query.startswith('http://data.judaicalink.org/data/ep/'):
        print("Query is a valid Entity Page!")
        print("Corresponding JudaicaLink resources:")
        pprint.pprint(ep_index[query])
    else:
        try:
            res = ep_inv_index[query]
            print("Query is a valid JudaicaLink resource!")
            print("Corresponding Entity Page:")
            print(res)
        except KeyError:
            print("Query is neither an Entity Page nor a valid JudaicaLink resource.")
```

```python
query_ep('http://data.judaicalink.org/data/ep/1045778', ep_index, ep_inv_index)
```
