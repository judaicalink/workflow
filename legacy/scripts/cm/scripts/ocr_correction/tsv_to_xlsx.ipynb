```python
import csv, os, pprint
from xlsxwriter.workbook import Workbook
path_data = "/data/cm/"
path = "/home/rovera/cm/"
```

```python
def tsv_to_xlsx(tsv_file, xlsx_file):

    workbook = Workbook(xlsx_file)
    worksheet = workbook.add_worksheet()

    tsv_reader = csv.reader(open(tsv_file, 'rt', encoding='utf8'), delimiter='\t')

    for row, data in enumerate(tsv_reader):
        worksheet.write_row(row, 0, data)
    

    workbook.close()
    print("Done!")
```

```python
tsv_file = path_data+"data/ocr_correction/empty/annotation_set_2.csv"
xlsx_file = path_data+"data/ocr_correction/empty/annotation_set_2.xlsx"
tsv_to_xlsx(tsv_file, xlsx_file)
```
