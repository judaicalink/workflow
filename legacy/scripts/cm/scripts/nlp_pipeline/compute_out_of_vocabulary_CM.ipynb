```python
import os, pprint, json, pickle
import spacy, gensim
path_data = "/data/cm/"
path = "/home/rovera/cm/"
```

```python
we = gensim.models.KeyedVectors.load_word2vec_format(path_data+'word_emb/cc.de.300.vec', binary=False)
```

```python
voc = pickle.load(open(path_data+"output/nlp_pipeline/word_count_CM/word_count_CM.pickle", 'rb'))
```

```python
oov = {}
for k,v in voc.items():
    try:
        we[k]
    except KeyError:
        oov[k] = v
```

```python
oov_sorted = sorted([(k,v) for k,v in oov.items()], key=lambda x:x[1], reverse=True)
```
