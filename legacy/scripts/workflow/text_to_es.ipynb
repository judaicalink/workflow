```python
import os, json, pprint
base = '/data/cm/output/'
```

```python
for journal in os.listdir(base+'text/'):
    
    print('Processing:', journal)
    
    path_struct = os.walk(base+'text/'+journal+'/')
    for dirpath, dirs, files in os.walk(base+'text/'+journal+'/'):
    
        if files != []:
            es_data = []
        
            #print(dirpath)
            path = dirpath.split(base+'text/')[1]
            pathname = '_'.join(path.split('/'))
        
            for file in files:
            
        
                # prepare index name for es
                filename = file.split('.')[0]
            
                path_based_id = pathname + '__' + filename
        
                # get text
                text = open(dirpath+'/'+file, 'r', encoding='utf-8').read()
        
                # prepare json data
                es_data.append(json.dumps({'index': {'_id': path_based_id, '_index': 'cm'}}, ensure_ascii=False))
                es_data.append(json.dumps({'text': text}, ensure_ascii=False))
        
            # create output directory path
            dirout = '/data/cm/output/es/'+path
            if not os.path.exists(dirout):
                os.makedirs(dirout)
        
            # we create a file for each issue (heft), the file contains textual data divided by page
            with open(dirout+'/'+pathname+'.json', "w", encoding="utf-8") as write_file:
                write_file.write("\n".join(es_data))
            
```

```python
size = 0
for journal in os.listdir(base+'text/'):
    
    #print('Processing:', journal)
    
    path_struct = os.walk(base+'text/'+journal+'/')
    for dirpath, dirs, files in os.walk(base+'text/'+journal+'/'):
        for file in files:
            file_size = os.path.getsize(dirpath+'/'+file)
            size += file_size
print("Total size:", round((((size/1024)/1024)/1024), 2), 'GB')
```
