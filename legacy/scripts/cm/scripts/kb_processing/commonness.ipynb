```python
# compute commonness of each entity e as target for a given mention m: P(e|m)
# a = number of times entity e is used as a link destination for m
# b = total number of times m appears as a link
# commonness = a/b
# output e: {m1: comm, m2: comm, ....}

import os, pprint, sys, time, pickle, json
from bs4 import BeautifulSoup as bs
root = "/data/cm/"
root_data = root+"data/"
```

```python
e2sf = pickle.load(open(root+'data/wiki/de/entity_to_sfs_count.pickle', 'rb'))
sf2e = pickle.load(open(root+'data/wiki/de/sf_to_entities_count.pickle', 'rb'))
```

```python
commonness = {}

for e, mentions in e2sf.items():
    
    commonness[e] = {}
    
    for m, count in mentions.items():
        a = count
        b = sum(sf2e[m].values())
        
        comm = a/b
        
        commonness[e][m] = comm
```

```python
with open(root+'data/wiki/de/commonness.pickle', 'wb') as outfile:
    pickle.dump(commonness, outfile)
```
