```python
# create an mapping index for all entities extracted from wikipedia
# the mapping is intended to be used for limiting the size of the entity to entity dictionary
# input: list of entities (hrefs) extracted from wikipedia
# output: {entity: id} and {id: entity} dictionaries
import os, json, pprint, pickle, os, time
root = "/data/cm/"
root_data = root+"data/"
```

```python
entities = pickle.load(open(root+'data/wiki/de/entity_to_sfs.pickle', 'rb'))
entities = list(entities.keys())
```

```python
e2i = {} # entity to index dict
for i in range(len(entities)):
    e2i[entities[i]] = i
```

```python
len(e2i)
```

```python
i2e = {v:k for k,v in e2i.items()}
```

```python
with open(root+'data/wiki/de/e2i.pickle', 'wb') as outfile:
    pickle.dump(e2i, outfile)
```

```python
with open(root+'data/wiki/de/i2e.pickle', 'wb') as outfile:
    pickle.dump(i2e, outfile)
```
