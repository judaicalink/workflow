```python
# build partials entity to entity
import os, json, pprint, pickle, os, time
root = "/data/cm/"
root_data = root+"data/"
```

```python
for file in os.listdir(root+'output/wiki/entity_to_entity/partials/'):
    start = time.time()
    
    if not file.endswith('checkpoints'):
        
    
        e2e = {}
    
        print('Processing', file, end=' ')
        with open(root+'output/wiki/entity_to_entity/partials/'+file, 'rb') as f:
            data = pickle.load(f)
            
            for bag in data:
                for a in bag: # current target entity in the bag
                
                    if a not in e2e.keys():
                        e2e[a] = {}
                
                    for b in bag:
                    
                        if a != b:
                        
                            if b not in e2e[a].keys():
                                e2e[a][b] = 1
                            else:
                                e2e[a][b] += 1
                                
        e2e = {k: {k1: v1 for k1, v1 in v.items() if v1 > 1} for k,v in e2e.items()}
        with open(root+'output/wiki/entity_to_entity/partials_dict/'+file, 'wb') as outfile:
            pickle.dump(e2e, outfile)
        end = time.time()
        print(str(round(end-start, 2)))
```
