```python
import os, json, pickle, pprint, time
```

```python
cm_entities = pickle.load(open('/home/rovera/cm/scripts/generators/generator_all_entities/cm_tagme_resource_reference_data_06_04.pickle', 'rb'))
```

```python
entities = {}
for entity in cm_entities:
    label = entity['dbpedia'].lstrip('http://dbpedia.org/resource/')
    if label not in entities:
        entities[label] = 0
    entities[label] += 1
```

```python
sorted_entities = sorted([(k,v) for k,v in entities.items()], key=lambda x:x[1], reverse=True)
```
