```python
classified_resources = {}

c = 0
for mention in data:
    resource = mention['resource']
    resource_label = resource.lstrip('http://de.dbpedia.org/resource/')
    cats = set(classes[resource_label]['cats'])
    if len(cats.intersection(blacklist)) == 0:
        if len(cats.intersection(pers)) > 0:
            classified_resources[resource] = 'http://dbpedia.org/ontology/Person'
        elif len(cats.intersection(locs)) > 0:
            classified_resources[resource] = 'http://dbpedia.org/ontology/Location'
        elif len(cats.intersection(orgs)) > 0:
            classified_resources[resource] = 'http://dbpedia.org/ontology/Organisation'
print(len(classified_resources))
```

```python
# generate rdf dataset
g = Graph()
g.bind('rdf', RDF)
for resource, dbp_class in classified_resources.items():
    resource_uri = URIRef(resource)
    dbp_class_uri = URIRef(dbp_class)
    g.add((resource_uri, RDF.type, dbp_class_uri))
print(len(g))
```

```python
g.serialize(destination="/data/cm/output/generators/CM_Dataset/cm_categorization.ttl", format="ttl")
```
