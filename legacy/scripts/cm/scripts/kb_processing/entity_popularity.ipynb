```python
# compute entity popularity as the number of unique Wikipedia documents an entity appears in
# input: entity unique occurrence count
# output: entity popularity (dict)

import os, pprint, sys, time, pickle, json
root = "/data/cm/"
root_data = root+"data/"
```

```python
docs = 2316034
e_pop = {}
e_unique = pickle.load(open(root+'data/wiki/de/unique_entity_count.pickle', 'rb'))
```

```python
for e, count in e_unique.items():
    e_pop[e] = count/docs*100
```

```python
with open(root+'data/wiki/de/entity_popularity.pickle', 'wb') as outfile:
    pickle.dump(e_pop, outfile)
```
