```python
def group_by_journal(cm_entities):
    occ_by_journal = {}
    for mention in cm_entities:
        if mention['resource'] not in occ_by_journal:
            occ_by_journal[mention['resource']] = {}
        if mention['journal_id'] not in occ_by_journal[mention['resource']]:
            occ_by_journal[mention['resource']][mention['journal_id']] = 0
        occ_by_journal[mention['resource']][mention['journal_id']] += 1
    return occ_by_journal

def get_occ_by_journal(resource_uri, occ_by_journal, id_to_journal):
    sorted_occ = sorted([(k,v) for k,v in occ_by_journal[resource_uri].items()], key=lambda x:x[1], reverse=True)
    for journal in sorted_occ:
        print(id_to_journal[journal[0]], journal[1])
```

```python
occ_by_journal = group_by_journal(cm_entities)
```
