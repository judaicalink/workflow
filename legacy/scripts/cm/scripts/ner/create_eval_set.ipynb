```python
# collect 50 random pages from the test dataset
# create an evaluation file with all the entries
import os, json, pprint
import random
```

```python
# collect 50 random pages
entities = json.load(open('/data/cm/output/ner/entities_test.json', 'r', encoding='utf-8'))
```

```python
pages = list(set(el['page'] for el in entities))
random.shuffle(pages)
pages = pages[:22]
```

```python
# collect entities
#pprint.pprint(pages)
```

```python
eval_set = {page: [] for page in pages}
for mention in entities:
    page = mention['page']
    if page in eval_set.keys():
        eval_mention = mention
        del eval_mention['page']
        eval_mention['is_ne'] = -1
        eval_mention['is_label'] = -1
        eval_mention['is_complete'] = -1
        eval_mention['word_error'] = -1
        eval_set[page].append(eval_mention)
```

```python
with open('/data/cm/output/ner/evaluation_set.json', 'w') as outfile:
    json.dump(eval_set, outfile, ensure_ascii=False, indent=2)
```
