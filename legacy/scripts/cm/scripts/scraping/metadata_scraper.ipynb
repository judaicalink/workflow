```python
import os, pprint, requests, urllib
from bs4 import BeautifulSoup as bs
import pickle, time, json
root = '/data/cm/'
```

```python
journals = pickle.load(open(root+'data/journal_metadata/journal_list.pickle', 'rb'))
journals_meta = json.load(open(root+'data/journal_metadata/journal_metadata_title_lang.json', 'r', encoding="utf-8"))
```

```python
meta_names = {
    'j_title': 'mods_titleInfoTitleNotType',
    'j_lang': 'mods_languageLanguageTerm',
    }
```

```python
for vlid in journals_meta.keys():
    
    if journals_meta[vlid]['title'] == '' or journals_meta[vlid]['lang'] == '':
    
        print('Processing', vlid, end="...")
    
        j_title, j_lang = '', ''
    
        try:
            webpage = requests.get('http://sammlungen.ub.uni-frankfurt.de/cm/periodical/titleinfo/'+vlid, timeout=2.0)
            print(webpage.status_code)

            webpagestr = webpage.text
    
            soup = bs(webpagestr, 'lxml')
    
        
    
            try:
                j_title = soup.find('tr', {'id': 'mods_titleInfoTitleNotType'}).find('div').string
                j_lang = soup.find('tr', {'id': 'mods_languageLanguageTerm'}).find('td', {'class': 'value '}).string
            except AttributeError as ae:
                print(ae)
    
        except requests.exceptions.Timeout as t:
            print(t)
    
        if journals_meta[vlid]['title'] == '':
            if j_title is None:
                j_title = ""
            journals_meta[vlid]['title'] = j_title
        
        if journals_meta[vlid]['lang'] == '':
            if j_lang is None:
                j_lang = ""
            journals_meta[vlid]['lang'] = j_lang
    
        time.sleep(10.0)
```

```python
json.dump(journals_meta, open(root+'data/journal_metadata/journal_metadata_title_lang.json', 'w', encoding="utf-8"), indent=2, ensure_ascii=False)
```
