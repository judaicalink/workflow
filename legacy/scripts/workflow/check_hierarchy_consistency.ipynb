```python
# Count how many journals follow the canonical structure journal --> volume --> issue
import os, json, pprint
root_data = '/data/cm/'
root = '/home/rovera/cm/scripts/'
```

```python
j_structure = {}

for journal in os.listdir(root_data+'xml/'):
    
    j_structure[journal] = 0
    
    for volume in os.listdir(root_data+'xml/'+journal+'/'):
        
        if os.path.isfile(root_data+'xml/'+journal+'/'+volume):
            j_structure[journal] = 1
            break
        else:
        
            for issue in os.listdir(root_data+'xml/'+journal+'/'+volume+'/'):
            
                if os.path.isfile(root_data+'xml/'+journal+'/'+volume+'/'+issue):
                    j_structure[journal] = 2
                    break
                else:
                    for file in os.listdir(root_data+'xml/'+journal+'/'+volume+'/'+issue+'/'):
                        if os.path.isfile(root_data+'xml/'+journal+'/'+volume+'/'+issue+'/'+file):
                            j_structure[journal] = 3
                        elif os.path.isdir(root_data+'xml/'+journal+'/'+volume+'/'+issue+'/'+file):
                            j_structure[journal] = 4
                        break
                        
```

```python
count = {1: 0, 2: 0, 3: 0, 4: 0}
```

```python
for k,v in j_structure.items():
    count[v] += 1
print(count)
```
