```python
import os, json, pprint
```

```python
data = json.load(open('/data/cm/data/entity_classification/entity_classification.json', 'r', encoding='utf-8'))
```

```python
locs = tuple([
    "http://dbpedia.org/ontology/Place",
      "http://dbpedia.org/ontology/Location",
      "http://schema.org/Place"
])
orgs = tuple([
    "http://schema.org/Organization",
    "http://dbpedia.org/ontology/Organisation"
])
pers = tuple([
    "http://xmlns.com/foaf/0.1/Person",
    "http://dbpedia.org/ontology/Person",
    "http://schema.org/Person"
])
events = tuple([
    "http://schema.org/Event",
    "http://dbpedia.org/ontology/Event"
])
```

```python
loc, org, per, event = set([]), set([]), set([]), set([])
loc_uris, org_uris, per_uris, event_uris = set([]), set([]), set([]), set([])
for entity, val in data.items():
    for cat in val['cats']:
        if cat in locs:
            loc.add(entity)
            for uri in val['cats']:
                loc_uris.add(uri)
        elif cat in orgs:
            org.add(entity)
            for uri in val['cats']:
                org_uris.add(uri)
        elif cat in pers:
            per.add(entity)
            for uri in val['cats']:
                per_uris.add(uri)
        elif cat in events:
            event.add(entity)
            for uri in val['cats']:
                event_uris.add(uri)
```

```python
loc, org, per, event = list(loc), list(org), list(per), list(event)
```
