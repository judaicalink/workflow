```python
import os, json, pprint
```

```python
ev_set = json.load(open('/home/rovera/cm/scripts/ner/evaluation_set_annotated.json', 'r', encoding='utf-8'))
ev_set = [mention for page in ev_set.values() for mention in page]
print("Size of evaluation set:", len(ev_set))
```

```python
# Global precision (excluding MISC category)
tp, fp = 0, 0
for m in ev_set:
    if m['is_ne'] == 1 and m['is_label'] == 1:
        tp += 1
    else:
        fp += 1
p = tp/(tp+fp)
print("Global Precision:", p)
```

```python
# Precision by label
labels = {'PER': 0, 'LOC': 0, 'ORG': 0}
tp, fp = 0, 0
for l in labels:
    tp, fp = 0, 0
    for m in ev_set:
        if m['lab'] == l:
            if m['is_ne'] == 1 and m['is_label'] == 1:
                tp += 1
            else:
                fp += 1
    p = tp/(tp+fp)
    labels[l] = p
print('Precision by label:')
pprint.pprint(labels)
```

```python
# Global percentage of correctly detected and classified entities
corr_det_and_class = len([m for m in ev_set if m['is_ne'] == 1 and m['is_label'] == 1])
print("Global percentage of correctly detected and classified entities", corr_det_and_class/len(ev_set))
```

```python
# Percentage RECOGNITION
recognized = len([m for m in ev_set if m['is_ne'] == 1])
print("Percentage of correctly RECOGNIZED entities:", recognized/len(ev_set))
```

```python
# Percentage CLASSIFICATION
classified = len([m for m in ev_set if m['is_label'] == 1 and m['is_ne'] == 1])
total = len([m for m in ev_set if m['is_ne'] == 1])
print("Percentage of correctly CLASSIFIED entities:", classified/total)
```

```python
# Percentage of detected entities with spelling errors
e_with_errors = len([m for m in ev_set if m['word_error'] == 1])
print("Percentage of detected entities with spelling errors", e_with_errors/len(ev_set))
```

```python
# Percentage of misdetected entities
misdetected = len([m for m in ev_set if m['is_ne'] == 0])
print("Percentage of misdetected entities:", misdetected/len(ev_set))
```

```python
# Percentage of misclassified entities
misclassified = len([m for m in ev_set if m['is_ne'] == 1 and m['is_label'] == 0])
total = len([m for m in ev_set if m['is_ne'] == 1])
print("Percentage of misclassified entities:", misclassified/total)
```
