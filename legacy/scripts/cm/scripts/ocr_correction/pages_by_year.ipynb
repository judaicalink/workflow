```python
# using quick metadata reference, create a list of pagenames (files) sorted by decade
# this resource will be useful for any sampling of the CM corpus that requires temporal information

import os, pprint, json, pprint, pickle
path_data = "/data/cm/"
path = "/home/rovera/cm/"
```

```python
fname_to_meta = json.load(open(path_data+"data/metadata/short_metadata/filename_to_meta.json", 'r', encoding='utf-8'))
```

```python
decades = {y: [] for y in range(1760, 2020, 10)}
```

```python
for fname, val in fname_to_meta.items():
    
    if fname_to_meta[fname]['pdf'] is True: # !!!!
    
        try:
    
            year = int(fname_to_meta[fname]['y'])
    
            for y in decades.keys():
                if year >= y and year <= y+9:
                    decades[y].append(fname)
                
        except ValueError:
            pass
```

```python
pprint.pprint([(k, len(v)) for k,v in decades.items()])
```

```python
with open(path_data+"data/metadata/short_metadata/pages_by_year.pickle", 'wb') as outfile:
    pickle.dump(decades, outfile)
```

```python
with open(path_data+"data/metadata/short_metadata/pages_by_year.json", 'w') as outfile:
    json.dump(decades, outfile)
```
