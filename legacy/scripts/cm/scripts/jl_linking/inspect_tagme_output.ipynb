```python
import os, json, pprint, pickle
root_data = '/data/cm/'
```

```python
surfaces = {}

for folder in os.listdir(root_data+'output/linker/entities/'):
    for file in os.listdir(root_data+'output/linker/entities/'+folder+'/'):
        tagme_output = json.load(open(root_data+'output/linker/entities/'+folder+'/'+file, 'r', encoding="utf-8"))
        
        for match in tagme_output:
            spot = match['spot']
            if spot not in surfaces:
                surfaces[spot] = 0
            surfaces[spot] += 1
    print('|', end="")
```

```python
with open(root_data+'output/linker/surfaces_count.json', 'w', encoding="utf-8") as outfile:
    json.dump(surfaces, outfile, indent=2, ensure_ascii=False)
```

```python
print(len(surfaces))
surf_sorted = sorted([(k,v) for k,v in surfaces.items()], key=lambda x:x[1], reverse=True)
```

```python
with open(root_data+'output/linker/surfaces_count_sorted.pickle', 'wb') as outfile:
    pickle.dump(surf_sorted, outfile)
```

```python
wrong = ["wir für", "gemä", "ren", "a c h", "olle", "3gl"]
```

```python
check_wrong = {k: {'link_prob': 0, 'rho': 0, 'c': 0} for k in wrong}
for folder in os.listdir(root_data+'output/linker/entities/'):
    for file in os.listdir(root_data+'output/linker/entities/'+folder+'/'):
        tagme_output = json.load(open(root_data+'output/linker/entities/'+folder+'/'+file, 'r', encoding="utf-8"))
        
        for match in tagme_output:
            spot = match['spot']
            
            if spot in check_wrong:
                check_wrong[spot]['link_prob'] += match['link_probability']
                check_wrong[spot]['rho'] += match['rho']
                check_wrong[spot]['c'] += 1
    print('|', end="")
```
