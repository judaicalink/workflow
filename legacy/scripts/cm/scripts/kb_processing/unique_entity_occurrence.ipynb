```python
# compute unique entity occurrence in wikipedia (one count per document)
# output entity: number of wikipedia article entity appears in

import os, pprint, sys, time, pickle, json
root = "/data/cm/"
root_data = root+"data/"
```

```python
docs = 0
e_unique = {}

for file in os.listdir(root+'output/wiki/entity_to_entity/partials/'):
    
    if not file.endswith('checkpoints'):
    
        print(file)
    
        partial = pickle.load(open(root+'output/wiki/entity_to_entity/partials/'+file, 'rb'))
    
        for doc in partial:
            docs += 1
        
            bag = set(doc)
            for entity in bag:
                if not entity in e_unique:
                    e_unique[entity] = 0
                e_unique[entity] += 1
```
