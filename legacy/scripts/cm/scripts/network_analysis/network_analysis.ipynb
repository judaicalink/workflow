```python
def create_node_data(label_count, entity_classes):
    i = 0
    node2i = {}
    node_data = []
    for label, freq in label_count.items():
        #try:
        cat = label_classes[label]
        node = tuple([i, {'name': label, 'freq': freq, 'cat': cat}])
        if label not in node2i:
            node_data.append(node)
            node2i[label] = i
            i+= 1
        #except KeyError:
        #    pass
    i2node = {i: node for node, i in node2i.items()}
    return (node_data, node2i, i2node)
```

```python
def create_edge_data(label_corr_dict, node2i):
    edge_data = []
    for label, entity_id in node2i.items():
        correlations = label_corr_dict[label]
        for correlated_label, score in correlations.items():
            try:
                correlated_id = node2i[correlated_label]
                edge = tuple([entity_id, correlated_id, {'weight': score}])
                edge_data.append(edge)
            except KeyError:
                pass
    return edge_data
```

```python
node_data, node2i, i2node = create_node_data(label_count, label_classes)
```
