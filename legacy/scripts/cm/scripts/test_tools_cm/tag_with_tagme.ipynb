```python
import os, json, pprint, tagme, pickle
path_data = "/data/cm/"
path = "/home/rovera/cm/"
tagme.GCUBE_TOKEN = "082285fb-1487-41fc-a75d-77be6c3a3780-843339462"
```

```python
test_res = {}
```

```python
for file in os.listdir(path_data+"output/test_tools_cm/test_data/"):
    
    if not file.startswith('.'):

        print("Processing", file)
    
        res = []
    
        text = open(path_data+"output/test_tools_cm/test_data/"+file).read()
    
        ann = tagme.annotate(text, lang="de")
    
        for a in ann.get_annotations(0.25):
            a_json = {}
            a_json['b'] = a.begin
            a_json['e'] = a.end
            a_json['e_id'] = a.entity_id
            a_json['e_title'] = a.entity_title
            a_json['score'] = a.score
            a_json['mention'] = a.mention
            
            res.append(a_json)
        
        test_res[file.split('.')[0]] = res
```

```python
for k,v in test_res.items():
    print(k,v)
    break
```

```python
pickle.dump(test_res, open(path_data+"output/test_tools_cm/tagme_linking/tagging_output.pickle", 'wb'))
```
